## 06 — Model Selection
This notebook compiles all 9 model results across 3 vectorisation methods 
and selects the best model for optimisation.

### Purpose
A systematic comparison of all method + model combinations ensures the 
final model choice is evidence-based. The selection criteria are:
1. Highest accuracy and F1 score
2. Ability to provide probability scores (confidence %)
3. Ability to provide feature coefficients (for explainability)

```python
import pandas as pd

# ── Paste your actual results here after running scripts 03, 04, 05 ────────
results = {
    "CountVectorizer + LR":  {"accuracy": 0.9241, "f1": 0.9167},
    "CountVectorizer + NB":  {"accuracy": 0.9873, "f1": 0.9873},
    "CountVectorizer + SVM": {"accuracy": 0.9241, "f1": 0.9167},
    "N-grams + LR":          {"accuracy": 0.9241, "f1": 0.9167},
    "N-grams + NB":          {"accuracy": 0.9873, "f1": 0.9873},
    "N-grams + SVM":         {"accuracy": 0.9241, "f1": 0.9167},
    "TF-IDF + LR":           {"accuracy": 0.9873, "f1": 0.9870},
    "TF-IDF + NB":           {"accuracy": 0.9873, "f1": 0.9870},
    "TF-IDF + SVM":          {"accuracy": 0.9747, "f1": 0.9737},
}

df = pd.DataFrame(results).T
df = df.sort_values("accuracy", ascending=False)
print(df.to_string())
df.to_csv("outputs/model_comparison.csv")
print("\nSaved to outputs/model_comparison.csv")
```

### Key Findings
**Full results ranked by accuracy:**

| Method + Model | Accuracy | F1 |
|---|---|---|
| TF-IDF + LR | 98.73% | 98.70% |
| TF-IDF + NB | 98.73% | 98.70% |
| CountVectorizer + NB | 98.73% | 98.73% |
| N-grams + NB | 98.73% | 98.73% |
| TF-IDF + SVM | 97.47% | 97.37% |
| All LR/SVM (CV/Ngram) | 92.41% | 91.67% |

**Chosen model: TF-IDF + Logistic Regression**

Reasons:
- Tied highest accuracy (98.73%) with Naive Bayes
- Provides probability scores → confidence % for users
- Provides feature coefficients → explainable theme tags
- More interpretable than black-box models
